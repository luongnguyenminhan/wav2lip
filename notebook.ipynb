```python
# filelist
import os

data_path = "videos"
dataset = os.listdir(data_path)

for subset in dataset:
    with open(f"filelists/{subset}.txt", "w") as f:
        subset_path = os.path.join(data_path, subset)
        videos = os.listdir(subset_path)
        text = ""
        for file in videos:
            # file = file.split(".")[0]
            file_path = os.path.join(subset_path, file)
            text += file_path + "\n"
        f.writelines(text)
```
